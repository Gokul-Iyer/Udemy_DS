```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#generating random data
xs = np.random.uniform(-10,10,size=(1000,1))
zs = np.random.uniform(-10,10,size=(1000,1))
noise = np.random.uniform(-1,1,size=(1000,1))

generated_inputs = np.column_stack((xs,zs))
generated_targets = 2 * xs - 3 * zs + 5 + noise

#saving the data as npz file
np.savez("TF_intro",inputs = generated_inputs, targets = generated_targets)
```

```python
#Loading the npz file
training_data = np.load("TF_intro.npz")
```

```python
#model
input_size = 2
output_size = 1

model = tf.keras.Sequential([
    tf.keras.layers.Dense(
        output_size,
        kernel_initializer = tf.random_uniform_initializer(-0.1,0.1),
        bias_initializer = tf.random_uniform_initializer(-0.1,0.1)       
    )
    ])

custom_optimiser = tf.keras.optimizers.SGD(learning_rate=0.02)
model.compile(optimizer=custom_optimiser,loss="mean_squared_error")
model.fit(training_data["inputs"], training_data["targets"], epochs=100,verbose=2)
```
