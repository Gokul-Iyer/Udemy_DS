```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("3.01. Country clusters.csv")
df
```

```python
plt.scatter(df["Longitude"],df["Latitude"])
plt.show()
```

```python
x = df.iloc[:,1:3]
x
```

```python
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
results = kmeans.fit_predict(x)
results
```

```python
df["Clusters"] = results
df
```

```python
#plotting clusters
plt.scatter(df["Longitude"],df["Latitude"],c=df["Clusters"])
plt.show()
```
